- get lines which are strating with from and save in a list

```python
path='mbox-short.txt'
file=open(path)
for line in file:
    if line.startswith("From"):
        print(line)
```

```python
from_list=[]
for line in file:
    if line.startswith('From'):
        from_list.append(line)
```

```python
path='mbox-short.txt'
file=open(path)
from_list=[line for line in file if line.startswith('From')]
len(from_list)
```

- in above output we need to remove \n at the end

```python
list1=[]
for line in from_list:
    list1.append(line.rstrip('\n'))
```
